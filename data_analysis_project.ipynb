```python
sns.relplot(data=df,x='math score',y='reading score',hue='gender')
```

```python
sns.regplot(data=df,x='writing score',y='reading score')
```

```python
sns.scatterplot(data=df,x='writing score',y='reading score',hue='parental level of education')
```

```python
sns.boxenplot(data=df,x='writing score',y='reading score')
```

```python
sns.boxenplot(data=df,x='writing score',y='reading score',hue='gender')
```

```python
sns.displot(data=df,x='writing score',y='math score',hue='lunch',kind='hist')
```

```python
sns.displot(data=df,x='writing score',y='math score',hue='lunch',kind='kde')
```

```python
sns.factorplot(data=df,x='writing score',y='reading score',hue='lunch')
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='reading score',y='writing score',hue='gender',palette='Dark2',size=200,alpha=0.5)
```

```python
plt.figure(figsize=(15,10))
sns.rugplot(data=df,x='math score',y='reading score',height=0.8)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=df,x='math score',y='reading score',hue='gender',palette='Set2')
plt.legend(bbox_to_anchor=(1,0.5))
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(data=df,x='writing score',y='reading score',hue='lunch',palette='Set2')
plt.legend(bbox_to_anchor=(1,0.5))
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(data=df,x='writing score',y='reading score',hue='parental level of education',palette='cividis_r')
plt.legend(bbox_to_anchor=(1,0.5))
```

```python
plt.figure(figsize=(15,10))
sns.boxenplot(data=df,x='writing score',y='reading score',hue='test preparation course',palette='PiYG')
plt.legend(bbox_to_anchor=(1,0.5))
```

```python
sns.jointplot(data=df,x='writing score',y='reading score',hue='gender',kind='scatter')
```

```python
sns.pairplot(data=df,hue='reading score',diag_kind='kde',corner='True')
```

```python
sns.catplot(data=df,x='writing score',y='reading score',hue='gender',kind='box',row='test preparation course')
```
